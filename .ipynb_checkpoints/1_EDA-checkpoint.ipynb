<h1> 1. Business Problem </h1>

## 1.1 Description of the Challenge

Mercari, Japan’s biggest community-powered shopping app (like OLX in India) is a platform on which sellers can sell their used/refurbished goods and thus it connects buyers to the sellers.<br> <br>
In this competition, the seller puts up an item/bundle of items on the app for sale and gives the description of the items including the item name, description etc. and Mercari gives a price suggestion for how many dollars approximately can this item be sold online.<br><br>
So, basically, Mercari is offering pricing suggestions to sellers for the products the seller wishes to put up on sale on Mercari.<br>
In this competition, we will predict the sale price of a listing based on information a seller provides for this listing
<br>
<br>
<b> <u> CRUX : The seller can provide details about the product like text descriptions of their products, including details like product category name, brand name, and item condition and our goal is to build an algorithm that automatically suggests the right product prices. </u> </b>

<br>
The exact details that the seller can provide on the Mercari app are:

● item_name (Text): The title of the listing (IMP attribute)<br>
● item_condition (Ordinal): The condition of the items provided by the seller measured in numbers from 1-5 (worst-best)<br>
● category_name (Categorical): category of the listing like Men/Women/Electrical/Antiques<br>
● brand_name (Categorical): like Louis Vitton etc.<br>
● shipping (Binary): 1 if shipping fee is paid by seller and 0 by buyer<br>
● item_description (Text): The full description of the item (IMP attribute)<br>

(Note: Here, we aren't using images for predicting the price)
<br>
<br>
Given all above details, our ML algorithm has to predict:
Price (numerical, in USD): The price that the item was sold for <br>
The train data consists of 1.5 million rows and thus we've enough data at our hands<br> <br>
The main challenge in this problem is to capture the small details in the product description, brand name etc. as these could heavily affect the product price. For eg., An original pair of Adidas shoes can be resold for upto 2000 in India, while, we
can get a pair of "Abibas" shoes for around 200 in the local market. <br> <br>
The goal of the ML algorithm is to find out the nuances in the data (which humans can't find so accurately or have biases) that can lead to significant increase/decrease in a product's selling price.<br> <br>

<h2>1.3 Real world/Business Objectives and Constraints </h2>

1. The suggested price should be as close to the original price as possible. In any case, it should not be lower than what the buyer is willing to pay, otherwise it will be a loss for both the seller and Mercari Platform.<br>
2. We have some low latency constraints. The predicted price should be given in few minutes.<br>
3. Interpretability is partially important.

<h2> 2.2 Mapping the real world problem to an ML problem </h2>

<h3> 2.2.1 Type of Machine Leaning Problem </h3>

<p> It is a regression problem, for a given item that the seller uploads, we've to predict the appropriate price at which this could be sold to any buyer on the Mercari Platform </p>

<h3> 2.2.2 Performance Metric </h3>

The performance metric used here is <b> Root Mean Squared Log Error (RMSLE) </b> <br>

Credits: [Medium Blog](https://medium.com/analytics-vidhya/root-mean-square-log-error-rmse-vs-rmlse-935c6cc1802a)<br><br>
Note that in the formulation X is the predicted value and Y is the actual value <br><br>
Here's a very nice blog explaining in detail about this metric: https://medium.com/analytics-vidhya/root-mean-square-log-error-rmse-vs-rmlse-935c6cc1802a <br>
Important points about this metric are:
1. Robust to outliers <br>
2. Calculates relative error instead of absolute error <br>
3. RMSLE incurs a larger penalty for the underestimation of the Actual variable than the Overestimation <br>

Due to these properties, we can say that RMSLE is the best metric for our business problem as it meets most of our business requirements

## Item Name

#### Let's preprocess the name field by removing stopwords, special characters and then build worcloud on it

Code in this section is inspired from the Notebook: https://www.kaggle.com/thykhuely/mercari-interactive-eda-topic-modelling


```python
'''
References:
1. EDA on Mercari: https://www.kaggle.com/thykhuely/mercari-interactive-eda-topic-modelling
2. Removing speical characters: https://stackoverflow.com/questions/6323296/python-remove-anything-that-is-not-a-letter-or-number
'''

stop = set(stopwords.words('english'))

def tokenize(text):
    """
    sent_tokenize(): segment text into sentences
    word_tokenize(): break sentences into words
    """
    regex = re.compile('[' + re.escape(string.punctuation) + '0-9\\r\\t\\n]')
    
    # remove punctuation
    text = regex.sub(" ", text)
    
    #removing special characters
    text = re.sub('[\W_]+', ' ', text) 
    
    tokens_orig = [word_tokenize(s) for s in sent_tokenize(text)]
    
    tokens = []
    
    for token_by_sent in tokens_orig:
        tokens += token_by_sent
    
    tokens = list(filter(lambda t: t.lower() not in stop, tokens))
    
    filtered_tokens = [w for w in tokens if re.search('[a-zA-Z]', w)]
    
    filtered_tokens = [w.lower() for w in filtered_tokens if len(w)>=3]
        
    return filtered_tokens
```

```python
# apply the tokenizer into the item name column
df_train['unigram_tokens_name'] = df_train['name'].map(tokenize)
```

```python
list_unigrams = list(df_train['unigram_tokens_name'])
```

```python
#Ref: https://stackoverflow.com/questions/952914/how-to-make-a-flat-list-out-of-list-of-lists
list_unigrams_flattened = []
for sublist in list_unigrams:
    if sublist is not None:
        for item in sublist:
            list_unigrams_flattened.append(item)
```

```python
string_unigrams_name = ' '.join(list_unigrams_flattened)
```

### Length of project description

```python
stop_words = set(stopwords.words('english'))
```

```python
#Ref: https://www.kaggle.com/thykhuely/mercari-interactive-eda-topic-modelling

def wordCount(text):
    # convert to lower case and strip regex
    try:
         # convert to lower case and strip regex
        text = text.lower()
        regex = re.compile('[' + re.escape(string.punctuation) + '0-9\\r\\t\\n]')
        txt = regex.sub(" ", text)
        # remove words in stop words
        words = [w for w in txt.split(" ") if w not in stop_words and len(w) > 3]
        return len(words)
    except:
        return 0
```

```python
df_train['desc_length'] = df_train['item_description'].apply(lambda x: wordCount(x))
```

```python
plt.xlabel('Description Length', fontsize = 20)
plt.ylabel('Log Price', fontsize = 20)
plt.xticks(fontsize = 15)
plt.yticks(fontsize = 15)
plt.title('Description Length vs Log Price', fontsize = 20)
sns.lineplot(x='desc_length', y='log_price', data=df_train)
plt.show()
```

#### Observations:
1. Items with short description length tend to have a smaller price.
2. Items with higher price tend to have a larger description length but converse is not always true.
